If you forget eval() and only use no_grad(), your predictions become noisy and non-deterministic because dropout is still randomly zeroing activations (and BatchNorm misbehaves on small or single-example batches). This is a sneaky one — no error, just quietly worse results.

If you forget no_grad() and only use eval(), your predictions are correct, but you waste memory building a graph you'll never use, which can cause out-of-memory errors when evaluating on large inputs.
